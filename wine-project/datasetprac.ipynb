Load dataset:

```python
import pandas as pd
df = pd.read_csv("winemag-data_first150k.csv")
df.head()
```

```python
df.info()
```

Step 1: Loading  & cleaning  data

```python
# Keeping only needed columns
df = df[['description', 'points']]

#checking for null values
df = df.dropna()
df.head()
```

Understanding the data:

```python
df['points'].describe()
```

Converting  text to  numbers:

```python
# Import TF-IDF tool
# This converts text (words) into numbers that ML models can understand
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF object
# max_features=5000 means:
# 👉 only keep top 5000 most important words in the dataset
# 👉 helps reduce noise + improves performance
vectorizer= TfidfVectorizer(max_features=5000)

# Convert text column into numerical matrix
# fit_transform does TWO things:
# 1. "fit" → learn vocabulary (important words from all descriptions)
# 2. "transform" → convert each description into a numeric vector
X=vectorizer.fit_transform(df['description'])


# Target variable (what we want to predict)
# Here: wine rating (points)
y=df['points']

#just to see what data vectorizer created and what does x and y look like:

# WHAT is X?
# A matrix like:
# Number of rows = number of wines
# Number of columns = number of words (5000)
# Each row = one wine review
# Each column = one word
print(f"Shape of X (features): {X.shape}")


# WHAT is y?
#  Just the answer column:
# points = [90, 85, 88, 92, ...]
print(f"Shape of y (target): {y.shape}")


#Remember: X = features (TF-IDF vectors), y = target (points)
#Good features = things that help distinguish outcomes
#Unique words might be informative, but they are not always reliable signals
```

Train/Test split

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(#this is the function that distributes the data into training and testing sets.
    X, y, test_size=0.2, random_state=42
)#Training = learning
#Testing = proving understanding on unseen data
```

Train model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)#train the linear regression model on the training data (X_train, y_train). Fit = make the line best match the data points
```

Evaluating the Model:

```python
from sklearn.metrics import mean_squared_error
predictions=model.predict(X_test)

mse=mean_squared_error(y_test, predictions)
print("Mean squared error is :", mse)
```

Testing Manualyy:

```python
sample = ["Heitz has made this stellar rosé from the rare Grignolino grape since 1961. Ruby grapefruit-red, it's sultry with strawberry, watermelon, orange zest and salty spice flavor, highlighted with vibrant floral aromas."]

sample_vec = vectorizer.transform(sample)#This converts the sample review into a TF-IDF vector using the same vectorizer we used for training. It ensures that the sample is represented in the same way as the training data, allowing the model to make a prediction based on the learned patterns.

print(model.predict(sample_vec))#This line uses the trained linear regression model to predict the wine rating (points) for the sample review. The model takes the TF-IDF vector of the sample review as input and outputs a predicted rating based on the patterns it learned during training.
#The higher the predicted rating, the better the model thinks the wine is based on the review's content.it means that th model thinks the wine is pretty good based on the review's content.
```
